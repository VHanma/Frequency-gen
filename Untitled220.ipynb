```python
import json
import math
import wave
import hashlib
from pathlib import Path
from dataclasses import dataclass, asdict
import tkinter as tk
from tkinter import filedialog, simpledialog, messagebox

import numpy as np
from PIL import Image, ImageDraw

# ============================================================
# STACKED HYPOTHETICAL MORPHOGENETIC ENCODER v5
# ------------------------------------------------------------
# New ideas added from the uploaded Gariaev material:
#   1) Two-program model:
#        - geometry program
#        - instruction-text program
#   2) Context resolver:
#        - ambiguous candidate states
#        - full-sequence resolution
#   3) Dual memory:
#        - short-term state buffer
#        - long-term blueprint buffer
#   4) Nonlocal address graph
#   5) Phantom persistence layer
#   6) Build vs maintenance mode
#
# This is a speculative signal-art / pattern-encoding system.
# ============================================================

DEFAULT_SAMPLE_RATE = 48000
DEFAULT_DURATION_SEC = 72.0

BANDS = {
    "foundation": (300.0, 900.0),
    "bridge":     (1700.0, 1900.0),
    "carrier":    (2400.0, 2600.0),
    "overdrive":  (3600.0, 3800.0),
}

DEFAULT_WEIGHTS = {
    "geometry": 0.22,
    "instruction": 0.18,
    "levin": 0.20,
    "gariaev": 0.20,
    "scalar": 0.10,
    "rife": 0.10,
}

PHASE_CYCLE = np.array([0, 45, 90, 135, 180, 225, 270, 315], dtype=np.float64)


# -----------------------------
# Utility
# -----------------------------
def clamp(x, lo, hi):
    return max(lo, min(hi, x))


def normalize(arr, eps=1e-12):
    arr = np.asarray(arr, dtype=np.float64)
    mn = np.min(arr)
    mx = np.max(arr)
    if mx - mn < eps:
        return np.zeros_like(arr)
    return (arr - mn) / (mx - mn)


def smooth1d(arr, win):
    arr = np.asarray(arr, dtype=np.float64)
    if win <= 1:
        return arr.copy()
    kernel = np.ones(win, dtype=np.float64) / win
    return np.convolve(arr, kernel, mode="same")


def stereo_from_mono_and_phase(mono, phase_deg=0.0, side_amt=0.15, mod=None):
    mono = np.asarray(mono, dtype=np.float64)
    if mod is None:
        mod = np.ones_like(mono)
    phase = np.deg2rad(phase_deg)
    side = np.sin(np.linspace(0.0, 2.0 * np.pi * (0.7 + abs(math.sin(phase))), len(mono)))
    side = side * mod
    left = mono * (1.0 - side_amt) + side_amt * side
    right = mono * (1.0 - side_amt) - side_amt * side
    return np.stack([left, right], axis=1)


def write_wav_stereo(path, stereo, sample_rate, master_norm=0.92):
    stereo = np.asarray(stereo, dtype=np.float64)
    peak = max(np.max(np.abs(stereo)), 1e-12)
    stereo = master_norm * stereo / peak
    pcm = (stereo * 32767.0).astype(np.int16)

    with wave.open(str(path), "wb") as wf:
        wf.setnchannels(2)
        wf.setsampwidth(2)
        wf.setframerate(sample_rate)
        wf.writeframes(pcm.tobytes())


def hash_text_to_unit(text):
    h = hashlib.sha256(text.encode("utf-8")).hexdigest()
    return int(h[:8], 16) / 0xFFFFFFFF


def safe_image_open(path, size=(256, 256), mode="L"):
    return Image.open(path).convert(mode).resize(size)


# -----------------------------
# Feature extraction
# -----------------------------
@dataclass
class ImageFeatureSummary:
    brightness_mean: float
    contrast_mean: float
    edge_density_mean: float
    symmetry_mean: float
    region_count: int


def extract_image_maps(image_path, size=(256, 256), grid=(8, 8)):
    img = safe_image_open(image_path, size=size, mode="L")
    arr = np.asarray(img, dtype=np.float64) / 255.0

    gx = np.zeros_like(arr)
    gy = np.zeros_like(arr)
    gx[:, 1:-1] = arr[:, 2:] - arr[:, :-2]
    gy[1:-1, :] = arr[2:, :] - arr[:-2, :]
    grad = np.sqrt(gx**2 + gy**2)
    grad_n = normalize(grad)

    flip = np.fliplr(arr)
    sym_local = 1.0 - normalize(np.abs(arr - flip))

    h, w = arr.shape
    gh, gw = grid
    bh = h // gh
    bw = w // gw
    features = []

    for r in range(gh):
        for c in range(gw):
            y0, y1 = r * bh, (r + 1) * bh
            x0, x1 = c * bw, (c + 1) * bw
            block = arr[y0:y1, x0:x1]
            eblock = grad_n[y0:y1, x0:x1]
            sblock = sym_local[y0:y1, x0:x1]
            features.append({
                "region": [r, c],
                "brightness": float(np.mean(block)),
                "contrast": float(np.std(block)),
                "edge_density": float(np.mean(eblock > 0.35)),
                "symmetry": float(np.mean(sblock)),
                "center_xy": [float((x0 + x1) / 2 / w), float((y0 + y1) / 2 / h)],
            })

    summary = ImageFeatureSummary(
        brightness_mean=float(np.mean([f["brightness"] for f in features])),
        contrast_mean=float(np.mean([f["contrast"] for f in features])),
        edge_density_mean=float(np.mean([f["edge_density"] for f in features])),
        symmetry_mean=float(np.mean([f["symmetry"] for f in features])),
        region_count=len(features),
    )

    return {
        "image_array": arr,
        "edge_map": grad_n,
        "symmetry_map": sym_local,
        "features": features,
        "summary": asdict(summary),
        "grid": grid,
        "size": size,
    }


def spiral_order(features):
    def key_fn(f):
        x, y = f["center_xy"]
        dx = x - 0.5
        dy = y - 0.5
        r = math.hypot(dx, dy)
        ang = math.atan2(dy, dx)
        return (round(r, 4), ang)
    return sorted(features, key=key_fn)


# -----------------------------
# Symbol/text map
# -----------------------------
def render_text_token_map(tokens, size=(256, 256)):
    canvas = Image.new("L", size, color=0)
    draw = ImageDraw.Draw(canvas)
    w, h = size

    for i, token in enumerate(tokens):
        u = hash_text_to_unit(token)
        v = hash_text_to_unit(token[::-1] + f"_{i}")
        cx = int((0.15 + 0.7 * u) * w)
        cy = int((0.15 + 0.7 * v) * h)
        rr = int(12 + 40 * hash_text_to_unit(token + "_r"))
        intensity = int(100 + 155 * hash_text_to_unit(token + "_i"))
        draw.ellipse((cx - rr, cy - rr, cx + rr, cy + rr), fill=intensity)

        ang = 2 * np.pi * hash_text_to_unit(token + "_a")
        dx = int(rr * 1.8 * np.cos(ang))
        dy = int(rr * 1.8 * np.sin(ang))
        draw.line((cx, cy, cx + dx, cy + dy), fill=intensity, width=2)

    return normalize(np.asarray(canvas, dtype=np.float64) / 255.0)


# -----------------------------
# Two-program engine
# -----------------------------
def build_geometry_program(image_maps):
    """
    Program A: geometric body plan / scheme.
    """
    arr = image_maps["image_array"]
    edge = image_maps["edge_map"]
    sym = image_maps["symmetry_map"]

    geometry_map = normalize(
        0.40 * arr +
        0.35 * edge +
        0.25 * sym
    )

    boundaries = normalize(edge)
    symmetry_lock = normalize(sym)

    return {
        "geometry_map": geometry_map,
        "boundary_map": boundaries,
        "symmetry_lock": symmetry_lock,
    }


def build_instruction_program(tokens, image_maps, symbol_image_path=None):
    """
    Program B: meaningful text/instruction layer.
    """
    base_arr = image_maps["image_array"]

    if symbol_image_path:
        sym_img = safe_image_open(symbol_image_path, size=base_arr.shape[::-1], mode="L")
        sym_arr = normalize(np.asarray(sym_img, dtype=np.float64) / 255.0)
    else:
        sym_arr = render_text_token_map(tokens, size=base_arr.shape[::-1])

    instruction_map = normalize(0.45 * base_arr + 0.55 * sym_arr)

    token_units = np.array([hash_text_to_unit(t) for t in tokens], dtype=np.float64) if tokens else np.array([0.5])
    token_freqs = 220 + 880 * token_units

    return {
        "instruction_map": instruction_map,
        "token_units": token_units.tolist(),
        "token_freqs": token_freqs.tolist(),
        "tokens": tokens,
    }


# -----------------------------
# Ambiguity / context resolver
# -----------------------------
def candidate_bands_for_feature(f):
    """
    Each region gets multiple candidate bands first.
    Context will resolve later.
    """
    b = f["brightness"]
    c = f["contrast"]
    e = f["edge_density"]
    s = f["symmetry"]

    scores = {
        "foundation": (1.0 - b) * 0.45 + s * 0.25 + c * 0.15 + (1.0 - e) * 0.15,
        "bridge":     b * 0.35 + s * 0.25 + (1.0 - e) * 0.25 + c * 0.15,
        "carrier":    c * 0.35 + e * 0.25 + b * 0.15 + s * 0.25,
        "overdrive":  e * 0.50 + c * 0.20 + (1.0 - s) * 0.20 + (1.0 - b) * 0.10,
    }
    ranked = sorted(scores.items(), key=lambda kv: kv[1], reverse=True)
    return ranked[:3]


def build_nonlocal_graph(features):
    """
    Create long-distance links:
    - bilateral mirror partner
    - center-near clusters
    - semantic similarity links
    """
    graph = {i: set() for i in range(len(features))}
    centers = np.array([f["center_xy"] for f in features], dtype=np.float64)

    for i, fi in enumerate(features):
        xi, yi = fi["center_xy"]
        mirror_target = np.array([1.0 - xi, yi], dtype=np.float64)
        dists = np.sum((centers - mirror_target) ** 2, axis=1)
        j = int(np.argmin(dists))
        if j != i:
            graph[i].add(j)
            graph[j].add(i)

        # semantic links by feature similarity
        sims = []
        for k, fk in enumerate(features):
            if k == i:
                continue
            dist = (
                abs(fi["brightness"] - fk["brightness"]) +
                abs(fi["contrast"] - fk["contrast"]) +
                abs(fi["edge_density"] - fk["edge_density"]) +
                abs(fi["symmetry"] - fk["symmetry"])
            )
            sims.append((dist, k))
        sims.sort(key=lambda x: x[0])
        for _, k in sims[:2]:
            graph[i].add(k)
            graph[k].add(i)

    return {k: sorted(list(v)) for k, v in graph.items()}


def resolve_context(features, graph):
    """
    Full-sequence contextual resolution:
    - start with top candidate
    - reweight by neighbors, symmetry, global means
    """
    global_b = np.mean([f["brightness"] for f in features])
    global_c = np.mean([f["contrast"] for f in features])
    global_e = np.mean([f["edge_density"] for f in features])
    global_s = np.mean([f["symmetry"] for f in features])

    resolved = []

    for i, f in enumerate(features):
        cands = candidate_bands_for_feature(f)
        band_scores = {name: score for name, score in cands}

        neighbor_idxs = graph.get(i, [])
        if neighbor_idxs:
            nb = np.mean([features[j]["brightness"] for j in neighbor_idxs])
            nc = np.mean([features[j]["contrast"] for j in neighbor_idxs])
            ne = np.mean([features[j]["edge_density"] for j in neighbor_idxs])
            ns = np.mean([features[j]["symmetry"] for j in neighbor_idxs])
        else:
            nb, nc, ne, ns = global_b, global_c, global_e, global_s

        # context shifts
        band_scores["foundation"] = band_scores.get("foundation", 0.0) + 0.10 * (1.0 - nb) + 0.06 * ns
        band_scores["bridge"]     = band_scores.get("bridge", 0.0)     + 0.08 * nb + 0.05 * ns
        band_scores["carrier"]    = band_scores.get("carrier", 0.0)    + 0.08 * nc + 0.08 * ne
        band_scores["overdrive"]  = band_scores.get("overdrive", 0.0)  + 0.12 * ne + 0.04 * (1.0 - ns)

        chosen = max(band_scores.items(), key=lambda kv: kv[1])[0]
        resolved.append({
            "index": i,
            "region": f["region"],
            "candidates": cands,
            "chosen_band": chosen,
            "neighbor_count": len(neighbor_idxs),
        })

    return resolved


# -----------------------------
# Memory layers
# -----------------------------
def build_memory_buffers(t, image_maps, geometry_program, instruction_program, mode="build"):
    """
    Short-term state memory + long-term blueprint memory.
    """
    geo_mean = float(np.mean(geometry_program["geometry_map"]))
    instr_mean = float(np.mean(instruction_program["instruction_map"]))
    sym_mean = image_maps["summary"]["symmetry_mean"]

    # long-term blueprint = slow, stable
    blueprint = (
        0.50
        + 0.20 * np.sin(2 * np.pi * (0.05 + 0.03 * geo_mean) * t)
        + 0.15 * np.sin(2 * np.pi * (0.08 + 0.03 * sym_mean) * t + 0.6)
    )
    blueprint = normalize(blueprint)
    blueprint = 0.4 + 0.6 * blueprint

    # short-term state = quicker, more current-condition dependent
    state = (
        0.50
        + 0.25 * np.sin(2 * np.pi * (0.14 + 0.06 * instr_mean) * t + 0.2)
        + 0.15 * np.sin(2 * np.pi * (0.22 + 0.04 * geo_mean) * t + 1.2)
    )
    state = normalize(state)
    state = 0.3 + 0.7 * state

    if mode == "maintenance":
        blend = 0.30 * state + 0.70 * blueprint
    else:
        blend = 0.55 * state + 0.45 * blueprint

    return {
        "state_buffer": state,
        "blueprint_buffer": blueprint,
        "blend_buffer": blend,
    }


# -----------------------------
# Geometry layer
# -----------------------------
def build_geometry_layer(t, geometry_program, memory_buffers, strength=1.0):
    gmean = float(np.mean(geometry_program["geometry_map"]))
    bmean = float(np.mean(geometry_program["boundary_map"]))
    smean = float(np.mean(geometry_program["symmetry_lock"]))
    mem = memory_buffers["blend_buffer"]

    mono = (
        0.55 * np.sin(2 * np.pi * (68 + 20 * gmean) * t) +
        0.30 * np.sin(2 * np.pi * (136 + 26 * bmean) * t + 0.7) +
        0.15 * np.sin(2 * np.pi * (272 + 30 * smean) * t + 1.4)
    )
    mono *= mem
    stereo = stereo_from_mono_and_phase(mono, phase_deg=180 * smean, side_amt=0.08, mod=mem)
    stereo *= strength / max(np.max(np.abs(stereo)), 1e-12)
    return stereo


# -----------------------------
# Instruction layer
# -----------------------------
def build_instruction_layer(t, instruction_program, memory_buffers, strength=1.0):
    freqs = instruction_program["token_freqs"]
    token_units = instruction_program["token_units"]
    mem = memory_buffers["state_buffer"]

    mono = np.zeros_like(t)
    for i, (f, u) in enumerate(zip(freqs, token_units)):
        env = 0.55 + 0.45 * np.sin(2 * np.pi * (0.10 + 0.03 * i) * t + 2 * np.pi * u)
        mono += (0.35 + 0.15 * u) * env * np.sin(2 * np.pi * f * t + 2 * np.pi * u)

    mono *= (0.75 + 0.25 * mem)
    stereo = stereo_from_mono_and_phase(mono, phase_deg=90, side_amt=0.16, mod=mem)
    stereo *= strength / max(np.max(np.abs(stereo)), 1e-12)
    return stereo


# -----------------------------
# Levin backbone
# -----------------------------
def build_levin_layer(t, image_maps, memory_buffers, strength=1.0):
    summary = image_maps["summary"]
    b = summary["brightness_mean"]
    c = summary["contrast_mean"]
    e = summary["edge_density_mean"]
    s = summary["symmetry_mean"]
    env = memory_buffers["blueprint_buffer"]

    base = (
        0.55 * np.sin(2 * np.pi * (72 + 18 * b) * t) +
        0.30 * np.sin(2 * np.pi * (144 + 40 * c) * t + 0.8) +
        0.15 * np.sin(2 * np.pi * (288 + 50 * s) * t + 1.3)
    )

    mono = strength * env * base
    stereo = stereo_from_mono_and_phase(mono, phase_deg=180 * s, side_amt=0.08, mod=env)
    return stereo


# -----------------------------
# Gariaev contextual layer
# -----------------------------
def band_pair_from_name(name, feature):
    lo, hi = BANDS[name]
    p1 = clamp(0.12 + 0.68 * feature["brightness"], 0.0, 1.0)
    p2 = clamp(0.12 + 0.68 * (0.5 * feature["contrast"] + 0.5 * feature["edge_density"]), 0.0, 1.0)
    f1 = lo + (hi - lo) * p1
    f2 = lo + (hi - lo) * p2
    if feature["symmetry"] > 0.7:
        f1, f2 = f2, f1
    return f1, f2


def build_gariaev_layer(t, features, resolved_context, graph, memory_buffers, strength=1.0):
    mono_l = np.zeros_like(t)
    mono_r = np.zeros_like(t)

    total_dur = t[-1] if len(t) > 1 else 1.0
    region_span = total_dur / max(1, int(len(features) * 0.45))
    mem = memory_buffers["blend_buffer"]

    for rc in resolved_context:
        i = rc["index"]
        f = features[i]
        band = rc["chosen_band"]
        f1, f2 = band_pair_from_name(band, f)

        start = (i * 0.33 * region_span) % max(total_dur, 1e-6)
        dur = 0.9 + 0.8 * clamp(0.4 * f["contrast"] + 0.3 * f["edge_density"] + 0.3 * f["symmetry"], 0.0, 1.0)
        local = np.clip((t - start) / dur, 0.0, 1.0)

        gate = np.where((t >= start) & (t <= start + dur), 1.0, 0.0)
        gate *= np.sin(np.pi * local) ** 2

        sweep = f1 + (f2 - f1) * local
        phase_base = 2 * np.pi * np.cumsum(sweep) / DEFAULT_SAMPLE_RATE

        pulse_rate = 2.0 + 9.0 * f["edge_density"]
        pulse = 0.55 + 0.45 * (0.5 + 0.5 * np.sign(np.sin(2 * np.pi * pulse_rate * t)))
        am = 0.75 + 0.25 * np.sin(2 * np.pi * (0.18 + 0.2 * f["brightness"]) * t + i * 0.17)

        sig = (
            0.62 * np.sin(phase_base + i * 0.11) +
            0.23 * np.sin(2.0 * phase_base + i * 0.07) +
            0.15 * np.sin(0.5 * phase_base + i * 0.19)
        )
        sig *= gate * pulse * am * (0.72 + 0.28 * mem)

        # nonlocal echo partners
        partner_mix = np.zeros_like(sig)
        for j in graph.get(i, []):
            fj = features[j]
            pf1, pf2 = band_pair_from_name(resolved_context[j]["chosen_band"], fj)
            psweep = pf1 + (pf2 - pf1) * local
            pphase = 2 * np.pi * np.cumsum(psweep) / DEFAULT_SAMPLE_RATE
            partner_mix += 0.12 * np.sin(pphase + j * 0.09)

        replica2_delay = int(DEFAULT_SAMPLE_RATE * (0.012 + 0.02 * f["symmetry"]))
        replica3_delay = int(DEFAULT_SAMPLE_RATE * (0.025 + 0.03 * f["contrast"]))

        sig2 = np.roll(sig + partner_mix, replica2_delay) * (0.42 + 0.18 * f["symmetry"])
        sig3 = np.roll(-(sig + 0.5 * partner_mix), replica3_delay) * (0.25 + 0.10 * f["contrast"])

        phase_deg = PHASE_CYCLE[i % len(PHASE_CYCLE)]
        phase_rad = np.deg2rad(phase_deg)

        mono_l += sig + np.cos(phase_rad) * sig2 + np.sin(phase_rad) * sig3
        mono_r += sig - np.cos(phase_rad) * sig2 - np.sin(phase_rad) * sig3

    stereo = np.stack([mono_l, mono_r], axis=1)
    stereo *= strength / max(np.max(np.abs(stereo)), 1e-12)
    return stereo


# -----------------------------
# Scalar layer
# -----------------------------
def build_scalar_layer(t, instruction_program, image_maps, memory_buffers, strength=1.0):
    mean_val = float(np.mean(instruction_program["instruction_map"]))
    edge_bias = float(np.mean(image_maps["edge_map"]))
    sym_bias = float(np.mean(image_maps["symmetry_map"]))
    mem = memory_buffers["state_buffer"]

    mono = (
        0.30 * np.sin(2 * np.pi * (600 + 400 * mean_val) * t) +
        0.22 * np.sin(2 * np.pi * (1300 + 500 * edge_bias) * t + 0.7) +
        0.16 * np.sin(2 * np.pi * (2100 + 700 * sym_bias) * t + 1.4)
    )
    mono *= (0.70 + 0.30 * mem)

    stereo = stereo_from_mono_and_phase(mono, phase_deg=90 + 180 * sym_bias, side_amt=0.18, mod=mem)
    stereo *= strength / max(np.max(np.abs(stereo)), 1e-12)
    return stereo


# -----------------------------
# Rife layer
# -----------------------------
def build_rife_layer(t, image_maps, memory_buffers, strength=1.0,
                     scan_bands=((700, 900), (1500, 1600), (2000, 2200), (4000, 5000)),
                     strike_period_sec=8.0,
                     strike_width_sec=1.65):
    edge = image_maps["summary"]["edge_density_mean"]
    contrast = image_maps["summary"]["contrast_mean"]
    strike_phase = np.mod(t, strike_period_sec)
    strike_gate = np.where(strike_phase < strike_width_sec, 1.0, 0.28)

    mono = np.zeros_like(t)
    for i, (lo, hi) in enumerate(scan_bands):
        sweep_rate = 0.08 + 0.03 * i + 0.04 * edge
        center = (lo + hi) / 2.0
        span = (hi - lo) / 2.0

        freq = center + span * np.sin(2 * np.pi * sweep_rate * t + i * 0.8)
        phase = 2 * np.pi * np.cumsum(freq) / DEFAULT_SAMPLE_RATE

        band_sig = np.sin(phase)
        pulse = 0.45 + 0.55 * (0.5 + 0.5 * np.sign(2 * np.pi * (2.5 + i + 2 * contrast) * t)) # Original had np.sin(2*np.pi*...) here, adjusted to match pulse logic
        band_sig *= strike_gate * pulse * (0.8 + 0.2 * memory_buffers["state_buffer"])
        mono += (0.38 - 0.06 * i) * band_sig

    stereo = stereo_from_mono_and_phase(mono, phase_deg=225, side_amt=0.10, mod=strike_gate)
    stereo *= strength / max(np.max(np.abs(stereo)), 1e-12)
    return stereo, strike_gate


# -----------------------------
# Phantom persistence layer
# -----------------------------
def build_phantom_layer(composite_prephantom, strength=0.18):
    """
    Creates return / afterimage / ghost behavior.
    """
    left = composite_prephantom[:, 0]
    right = composite_prephantom[:, 1]

    d1 = int(DEFAULT_SAMPLE_RATE * 0.35)
    d2 = int(DEFAULT_SAMPLE_RATE * 0.90)
    d3 = int(DEFAULT_SAMPLE_RATE * 1.80)

    ghost_l = (
        0.55 * np.roll(left, d1) +
        0.28 * np.roll(left, d2) -
        0.17 * np.roll(left, d3)
    )
    ghost_r = (
        0.55 * np.roll(right, d1) +
        0.28 * np.roll(right, d2) -
        0.17 * np.roll(right, d3)
    )

    # slow re-emergence
    t = np.arange(len(left), dtype=np.float64) / DEFAULT_SAMPLE_RATE
    reappear = 0.35 + 0.65 * (0.5 + 0.5 * np.sin(2 * np.pi * 0.045 * t + 0.8))

    stereo = np.stack([ghost_l * reappear, ghost_r * reappear], axis=1)
    stereo *= strength / max(np.max(np.abs(stereo)), 1e-12)
    return stereo


# -----------------------------
# Build / maintenance mode weighting
# -----------------------------
def mode_weights(mode):
    if mode == "maintenance":
        return {
            "geometry_gain": 0.80,
            "instruction_gain": 1.00,
            "levin_gain": 1.00,
            "gariaev_gain": 0.95,
            "scalar_gain": 1.00,
            "rife_gain": 0.75,
            "phantom_gain": 1.00,
        }
    return {
        "geometry_gain": 1.00,
        "instruction_gain": 0.90,
        "levin_gain": 1.00,
        "gariaev_gain": 1.00,
        "scalar_gain": 0.90,
        "rife_gain": 0.90,
        "phantom_gain": 0.75,
    }


# -----------------------------
# Main render
# -----------------------------
def build_composite_v5(
    image_path,
    symbol_image_path=None,
    tokens=None,
    out_dir="encoder_v5_output",
    sample_rate=DEFAULT_SAMPLE_RATE,
    duration_sec=DEFAULT_DURATION_SEC,
    mode="build",
    weights=None,
):
    if tokens is None:
        tokens = ["repair", "cohere", "restore", "symmetry", "integrate"]
    if weights is None:
        weights = DEFAULT_WEIGHTS.copy()

    out_dir = Path(out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    n = int(sample_rate * duration_sec)
    t = np.arange(n, dtype=np.float64) / sample_rate

    image_maps = extract_image_maps(image_path, size=(256, 256), grid=(8, 8))
    features = spiral_order(image_maps["features"])

    geometry_program = build_geometry_program(image_maps)
    instruction_program = build_instruction_program(tokens, image_maps, symbol_image_path=symbol_image_path)

    graph = build_nonlocal_graph(features)
    resolved_context = resolve_context(features, graph)

    memory_buffers = build_memory_buffers(
        t=t,
        image_maps=image_maps,
        geometry_program=geometry_program,
        instruction_program=instruction_program,
        mode=mode,
    )

    mw = mode_weights(mode)

    geometry_stereo = build_geometry_layer(t, geometry_program, memory_buffers, strength=1.0)
    instruction_stereo = build_instruction_layer(t, instruction_program, memory_buffers, strength=1.0)
    levin_stereo = build_levin_layer(t, image_maps, memory_buffers, strength=1.0)
    gariaev_stereo = build_gariaev_layer(t, features, resolved_context, graph, memory_buffers, strength=1.0)
    scalar_stereo = build_scalar_layer(t, instruction_program, image_maps, memory_buffers, strength=1.0)
    rife_stereo, strike_gate = build_rife_layer(t, image_maps, memory_buffers, strength=1.0)

    prephantom = (
        weights["geometry"]    * geometry_stereo    * mw["geometry_gain"] +
        weights["instruction"] * instruction_stereo * mw["instruction_gain"] +
        weights["levin"]       * levin_stereo       * mw["levin_gain"] +
        weights["gariaev"]     * gariaev_stereo     * mw["gariaev_gain"] +
        weights["scalar"]      * scalar_stereo      * mw["scalar_gain"] +
        weights["rife"]        * rife_stereo        * mw["rife_gain"]
    )

    phantom_stereo = build_phantom_layer(prephantom, strength=0.18 * mw["phantom_gain"])

    composite = prephantom + phantom_stereo
    composite[:, 0] = smooth1d(composite[:, 0], 5)
    composite[:, 1] = smooth1d(composite[:, 1], 5)

    write_wav_stereo(out_dir / "stacked_morphogenetic_encoder_v5_composite.wav", composite, sample_rate)
    write_wav_stereo(out_dir / "stem_geometry.wav", geometry_stereo, sample_rate)
    write_wav_stereo(out_dir / "stem_instruction.wav", instruction_stereo, sample_rate)
    write_wav_stereo(out_dir / "stem_levin.wav", levin_stereo, sample_rate)
    write_wav_stereo(out_dir / "stem_gariaev.wav", gariaev_stereo, sample_rate)
    write_wav_stereo(out_dir / "stem_scalar.wav", scalar_stereo, sample_rate)
    write_wav_stereo(out_dir / "stem_rife.wav", rife_stereo, sample_rate)
    write_wav_stereo(out_dir / "stem_phantom.wav", phantom_stereo, sample_rate)

    manifest = {
        "input_image": str(image_path),
        "symbol_image": None if symbol_image_path is None else str(symbol_image_path),
        "mode": mode,
        "sample_rate": sample_rate,
        "duration_sec": duration_sec,
        "weights": weights,
        "tokens": tokens,
        "bands_hz": BANDS,
        "image_summary": image_maps["summary"],
        "context_resolution": resolved_context,
        "nonlocal_graph": graph,
        "memory": {
            "blueprint_mean": float(np.mean(memory_buffers["blueprint_buffer"])),
            "state_mean": float(np.mean(memory_buffers["state_buffer"])),
            "blend_mean": float(np.mean(memory_buffers["blend_buffer"])),
        },
    }

    with open(out_dir / "stacked_morphogenetic_encoder_v5_manifest.json", "w", encoding="utf-8") as f:
        json.dump(manifest, f, indent=2)

    print(f"[DONE] Output folder: {out_dir.resolve()}")
    print("[DONE] Composite + stems + manifest written.")


# -----------------------------
# Simple GUI
# -----------------------------
def choose_file(title, filetypes):
    root = tk.Tk()
    root.withdraw()
    root.attributes("-topmost", True)
    path = filedialog.askopenfilename(title=title, filetypes=filetypes)
    root.destroy()
    return path if path else None


def ask_yes_no(title, question):
    root = tk.Tk()
    root.withdraw()
    root.attributes("-topmost", True)
    result = messagebox.askyesno(title, question)
    root.destroy()
    return result


def ask_string(title, prompt, initialvalue=""):
    root = tk.Tk()
    root.withdraw()
    root.attributes("-topmost", True)
    result = simpledialog.askstring(title, prompt, initialvalue=initialvalue)
    root.destroy()
    return result


def ask_float(title, prompt, initialvalue):
    root = tk.Tk()
    root.withdraw()
    root.attributes("-topmost", True)
    result = simpledialog.askfloat(title, prompt, initialvalue=initialvalue)
    root.destroy()
    return result


def main():
    image_path = choose_file(
        "Choose main photo",
        [("Image files", "*.png *.jpg *.jpeg *.bmp *.tif *.tiff *.webp"), ("All files", "*.* ")]
    )
    if not image_path:
        print("No main photo selected. Exiting.")
        return

    use_symbol = ask_yes_no("Symbol Overlay", "Do you want to add a symbol/sigil overlay image?")
    symbol_image_path = None
    if use_symbol:
        symbol_image_path = choose_file(
            "Choose symbol/sigil overlay image",
            [("Image files", "*.png *.jpg *.jpeg *.bmp *.tif *.tiff *.webp"), ("All files", "*.* ")]
        )

    tokens_raw = ask_string(
        "Tokens",
        "Enter tokens separated by commas:",
        initialvalue="repair, cohere, restore, symmetry, integrate"
    )
    if tokens_raw:
        tokens = [t.strip() for t in tokens_raw.split(",") if t.strip()]
    else:
        tokens = ["repair", "cohere", "restore", "symmetry", "integrate"]

    duration_sec = ask_float("Duration", "Enter duration in seconds:", initialvalue=72.0)
    if duration_sec is None:
        duration_sec = 72.0

    mode = ask_string("Mode", "Type mode: build or maintenance", initialvalue="build")
    if mode not in {"build", "maintenance"}:
        mode = "build"

    build_composite_v5(
        image_path=image_path,
        symbol_image_path=symbol_image_path,
        tokens=tokens,
        out_dir="encoder_v5_output",
        sample_rate=48000,
        duration_sec=duration_sec,
        mode=mode,
        weights=DEFAULT_WEIGHTS,
    )


if __name__ == "__main__":
    # main()
    pass # tkinter GUI is not suitable for Colab, use direct function calls instead
```
